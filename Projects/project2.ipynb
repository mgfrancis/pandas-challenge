```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file_to_load = "D:\\Repos1\\gwu-arl-data-pt-09-2020-u-c\\02-Homework\\04-Python-Pandas\\03-Pandas-Project\\Mini-Project-Part-2\\Resources\\2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
bootcamp2 = pd.read_csv(file_to_load, low_memory=False)
bootcamp2.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
extract2 = bootcamp2.iloc[:,[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
extract2.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
extract2.replace(to_replace = [0.0,1.0],value =['No','Yes'], inplace = True)
extract2.head()
```

```python
# Extract rows for only those who attended a bootcamp
attended2 = extract2.loc[(extract2['AttendedBootcamp'] == 'Yes')]
attended2.head()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
attendance = pd.DataFrame(attended2['BootcampName'].value_counts())

attendance.reset_index(inplace=True)
attendance.columns = ['BootcampName','AttendeesCount']
attendance.sort_values(by=['AttendeesCount'], inplace=True, ascending = False)
attendance.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
#recommend = attended2['BootcampRecommend'].count()
recommend = attended2.loc[attended2['BootcampRecommend'] =='Yes','BootcampName'].value_counts()
recommend = pd.DataFrame(recommend)
recommend.reset_index(inplace=True)
recommend.columns = ['BootcampName','RecommendCount']
recommend
```

```python
# Merge the two created data frames on the name of tbe bootcamp

bootcampmerge = pd.merge(attendance, recommend, on='BootcampName')
bootcampmerge
```

```python
# Calculate percentage of each bootcamp's students who are recommenders

yespct = round(bootcampmerge['RecommendCount'] / bootcampmerge['AttendeesCount'] * 100,2)
bootcampmerge['Recommend %'] = yespct
bootcampmerge
```

```python
# Export to excel and remove index
newdf = pd.DataFrame(bootcampmerge)
newdf.set_index('BootcampName')
```
