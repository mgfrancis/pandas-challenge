```python
# Read with pandas
bootcamp = pd.read_csv(file_to_load)#,encoding="utf-8")
```

```python
# Inspect all columns
bootcamp.head()
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111

extract = bootcamp.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
extract.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns

#extract['AttendedBootcamp'] = extract.loc[:,extract['AttendedBootcamp' == 0.0]].replace([0.0,1.0],['No','Yes'])
extract.replace(to_replace = [0.0,1.0],value =['No','Yes'], inplace = True)

#extract.replace([0.0,'No'])
extract.tail()
```

```python
# Calculate total number of respondents in survey
respondents = len(extract) 
rowcount= extract.count()
respondents, rowcount
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended = extract.loc[(extract['AttendedBootcamp']  == 'Yes')]
attended.head()
```

```python
attendees = attended['AttendedBootcamp'].count()
attendees
```

```python
# Calculate average age of attendees
# Calculate how many people attended a bootcamp
# Calculate how many attendees hold degrees
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
# Calculate percentage of respondents who attended a bootcamp
# Calculate percentage of respondents belonging to each gender
# Calculate percentage of attendees with a school degree
# Calculate average post-bootcamp salary
```

```python
# Calculate average age of attendees
aveage = round(attended['Age'].mean(),3)
aveage
```

```python
# Calculate how many people attended a bootcamp
attendees = attended['AttendedBootcamp'].count()
attendees
```

```python
# Calculate how many attendees hold degrees
#run valuecounts to check kinds of degrees
degree = attended.loc[attended['AttendedBootcamp'] == 'Yes','SchoolDegree'].value_counts()

#create list of degrees
wdegrees = ["bachelor's degree","master's degree (non-professional)", 'professional degree (MBA, MD, JD, etc.)',"associate's degree",'Ph.D.']

#create new table based on "isin" wdegrees list filter
newdf = attended[attended.SchoolDegree.isin(wdegrees)]
wdegreescount = newdf['AttendedBootcamp'].count()
wdegreescount
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
male = attended.loc[attended["Gender"] == "male","AttendedBootcamp"].count()
female = attended.loc[attended["Gender"] == "female","AttendedBootcamp"].count()
others = attended['Gender'].count() - male - female
male, female, others
```

```python
# Calculate percentage of respondents who attended a bootcamp
didattend = round((attendees / respondents)*100,2)

#formatting
didattend = "{:.2f}%".format(didattend)
didattend
```

```python
# Calculate percentage of respondents belonging to each gender
malepct = round((male/respondents) * 100,2)
femalepct = round((female/respondents) * 100,2)
othpct = round((others/respondents) * 100,2)
#formatting
malepct = "{:.2f}%".format(malepct)
femalepct = "{:.2f}%".format(femalepct)
othpct = "{:.2f}%".format(othpct)
malepct, femalepct, othpct
```

```python
# Calculate percentage of attendees with a school degree

degreepct = round((wdegreescount/attendees) * 100,2)
"{:.2f}%".format(degreepct)
```

```python
# Calculate average post-bootcamp salary
salary = attended['BootcampPostSalary'].mean()
#formatting
salary = "${:,.2f}".format(salary)
salary
```
