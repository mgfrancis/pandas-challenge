## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Establish bins for ages
agebins = [0,10,15,20,25,30,35,40,100]
binlabels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

#Categorize the existing players using the age bins.
purchase_data['Ages'] = pd.cut(purchase_data['Age'], agebins, labels=binlabels, include_lowest=True, right=False)


#Count per bin
less10 = purchase_data.loc[purchase_data['Ages'] == '<10', 'SN'].nunique()
less14 = purchase_data.loc[purchase_data['Ages'] == '10-14', 'SN'].nunique()
less19 = purchase_data.loc[purchase_data['Ages'] == '15-19', 'SN'].nunique()
less24 = purchase_data.loc[purchase_data['Ages'] == '20-24', 'SN'].nunique()
less29 = purchase_data.loc[purchase_data['Ages'] == '25-29', 'SN'].nunique()
less34 = purchase_data.loc[purchase_data['Ages'] == '30-34', 'SN'].nunique()
less39 = purchase_data.loc[purchase_data['Ages'] == '35-39', 'SN'].nunique()
abv40 = purchase_data.loc[purchase_data['Ages'] == '40+', 'SN'].nunique()

#Percentage per bin
pct10 = (less10/player_count)*100
pct14 = (less14/player_count)*100
pct19 = (less19/player_count)*100
pct24 = (less24/player_count)*100
pct29 = (less29/player_count)*100
pct34 = (less34/player_count)*100
pct39 = (less39/player_count)*100
pct40 = (abv40/player_count)*100

agesummary = pd.DataFrame({'Total Count': [less10, less14,less19, less24, less29, less34, less39, abv40],
                          'Percentage of Players': ["{:.2f}%".format(pct10), "{:.2f}%".format(pct14),
                           "{:.2f}%".format(pct19),"{:.2f}%".format(pct24),"{:.2f}%".format(pct29),"{:.2f}%".format(pct34),
                           "{:.2f}%".format(pct39),"{:.2f}%".format(pct40)]})
agesummary = agesummary.rename(index = {0 :'<10',1:'10-14',2:'15-19',3: '20-24',4:'25-29',5:'30-34',6:'35-39',7:'40+'})
agesummary
```
